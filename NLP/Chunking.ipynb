```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import RegexpParser
```

```python
nltk.download()
```

```python
dataset = "Taj Mahal is one of the world's most celebrated structures. It is a stunning symbol of rich Indian history"
```

```python
new_dataset = word_tokenize(dataset)
new_dataset
```

```python
# Applying POS tagging
postagging = pos_tag(new_dataset)
postagging
```

```python
# defining sequence of chunk
sequence_chunk = """
chunk:
{<NNPS>+}
{<NNP>+}
{<NN>+}"""
```

```python
# create object with regular expression
chunk = RegexpParser(sequence_chunk)
```

```python
chunk_result = chunk.parse(postagging)
print(chunk_result)
```
